```python
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('/home/jovyan/teaching_material/ECS765P/Coursework/PartA/total_txn/total_txn.csv')
```

```python
df1.columns = ['Date', 'Total Transactions each month']
df1.head()
```

```python
print('The graph showing number of transactions occurring every month between the start and end of the dataset: ')
plt.figure(figsize = (20,10))
plt.xticks(rotation=90)
plt.xlabel('Month/Year')
plt.ylabel('Total Transactions each month')
x=df1['Date']
y=df1['Total Transactions each month']
plt.bar(x,y,color='lightcoral')
```
